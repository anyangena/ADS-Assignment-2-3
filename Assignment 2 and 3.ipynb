```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
attr_data.Attrition.value_counts().plot(kind='bar', color='blue', alpha=.65)
plt.title('Attrition Breakdown')
```

```python
pd.crosstab(attr_data.Gender,attr_data.Attrition).plot(kind='bar')
plt.title('Attrition with respect to Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency of Attrition')
```

```python
pd.crosstab(attr_data.Department,attr_data.Attrition).plot(kind='bar')
plt.title('Attrition vs Department')
plt.xlabel('Department')
plt.ylabel('Frequency of attrition')
plt.xticks(rotation=90)
```

```python
pd.crosstab(attr_data.EducationField,attr_data.Attrition).plot(kind='bar')
plt.title('Attrition with respect to EducationField')
plt.xlabel('EducationField')
plt.ylabel('Frequency of Attrition')
```

```python
pd.crosstab(attr_data.OverTime,attr_data.Attrition).plot(kind='bar')
plt.title('Attrition with respect to OverTime')
plt.xlabel('OverTime')
plt.ylabel('Frequency of Attrition')
```
